```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder

warnings.filterwarnings('ignore')
```

```python
# Loading Dataset
data = pd.read_csv('book.csv')
```

```python
book_count = []
col_names = data.columns

# Loop through each column in the dataframe
for col_name in col_names:
    # Count the occurrences of the value 1 in the current column and append it to the book_count list
    book_count.append(data[col_name].value_counts()[1])

# Create a bar plot using the col_names as x-axis and the book_count as y-axis
plt.bar(col_names, book_count)
plt.xticks(rotation=45)  # Rotates the x-axis labels for better readability
plt.show()
```

```python
frequent_itemsets1 = apriori(data, min_support=0.1, use_colnames=True)
frequent_itemsets1 
```

```python
frequent_itemsets1 = apriori(data, min_support = 0.1, use_colnames=True)
frequent_itemsets1['length'] = frequent_itemsets1['itemsets'].apply(lambda x: len(x))
frequent_itemsets1
```

```python
rules1 = association_rules(frequent_itemsets1, metric="lift", min_threshold=0.5)
rules1
```

```python
rules1.sort_values('lift',ascending = False)
```

```python
plt.scatter(rules1['support'], rules1['confidence'], c=rules1['lift'], cmap='viridis', alpha=0.5)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Support vs Confidence colored by Lift')
plt.colorbar(label='Lift')
plt.show()
```

```python
rules2 = association_rules(frequent_itemsets1, metric="confidence", min_threshold=0.5)
rules2
```

```python
rules2 = rules2.sort_values(['confidence', 'lift'], ascending =[False, False])
rules2
```

```python
import seaborn as sns

# Define a color palette
my_palette = sns.color_palette('husl', n_colors=len(rules2))

plt.scatter(rules2['support'], rules2['confidence'], c=my_palette, alpha=0.5)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Support vs Confidence')
plt.show()
```
